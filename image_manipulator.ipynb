# Importing Libraries

```python
import numpy as np
from matplotlib.pyplot import imread
import matplotlib.pyplot as plt
```

## Loading image file

```python
img = imread("cat.jpg")
```

## Image pixel modifier parameter

```python
#By default the parameter is equal to one, which means that no changes are made on the pixels of the image.
# This paramter can be adjusted in order to se changes on the pixels of the image
a = 2
```

## Displaying the Orignal Image

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(img)
```

## Displaying modified image

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,2)
plt.imshow(img*a)
plt.show()
```
